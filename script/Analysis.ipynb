```python
import pandas as pd
```

```python
numbers3_data = pd.read_json("../preprocess_data3.json")
```

```python
numbers3_data
```

```python
numbers3_data = numbers3_data[numbers3_data["year"] > 2000]
```

```python
numbers3_data.set_index("id", inplace=True)
```

```python
[c for c in numbers3_data.columns if "prev_" not in c]
```

```python
X = numbers3_data.drop(['d1', 'd2', 'd3'], axis=1)
```

```python
Y = numbers3_data[['d1', 'd2', 'd3']]
```

```python
import lightgbm as lgb
```

```python
model = lgb.LGBMClassifier()
```

```python
X_train, X_test = X[:4000], X[4000:]
Y_train, Y_test = Y[:4000], Y[4000:]
```

```python
X_train = X_train.values
X_test = X_test.values
```

```python
Y_train = Y_train['d1'].values
Y_test = Y_test['d1'].values
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
Y_pred = model.predict_proba(X_test)
Y_pred_k = np.argmax(Y_pred, axis=1)
accu = accuracy_score(Y_test, Y_pred_k)
print(accu)
```
